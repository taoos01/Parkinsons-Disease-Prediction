Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Importing the dataset
parkinsons_data = pd.read_csv('parkinsons.csv')
```

```python
parkinsons_data.head()
```

```python
parkinsons_data.info()
```

# There are no missing values in the dataset

```python
# distribution of the terget variable
parkinsons_data['status'].value_counts()
```

## status
- 0 -> Healthy
- 1 -> Parkinsons positive

Dropping the name column

```python
parkinsons_data.drop(columns= 'name', axis= 1, inplace= True)
```

```python
# Grouping the data based on the target variable
parkinsons_data.groupby('status').mean()
```

Data preprocessing

Separating the features and the label

```python
X = parkinsons_data.drop(columns= 'status', axis= 1)
Y = parkinsons_data['status']
```

```python
print(X.shape, Y.shape)
```

Splitting the data into training data and the test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

Data Standardization because the values are not in a common range

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training

Support Vector Machine

```python
model = svm.SVC(kernel= 'linear')
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score of our model

```python
#Accuracy Score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print("Accuracy Score:" , round(training_data_accuracy * 100, 2))
```

```python
# Accuracy Score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction) * 100
```

```python
print("Accuracy Score:", test_data_accuracy)
```

Making a predictive system

```python
input_data = (119.99200,157.30200,74.99700,0.00784,0.00007,0.00370,0.00554,0.01109,0.04374,0.42600,0.02182,0.03130,0.02971,0.06545,0.02211,21.03300,0.414783,0.815285,-4.813031,0.266482,2.301442,0.284654
)

input_data_as_np_array = np.asarray(input_data)

reshaped_data = input_data_as_np_array.reshape(1, -1)

std_data = scaler.transform(reshaped_data)

prediction = model.predict(std_data)
print(prediction)

if(prediction[0] == 1):
    print("Parkinsons Disease positive")
else:
    print("Healthy")
```
